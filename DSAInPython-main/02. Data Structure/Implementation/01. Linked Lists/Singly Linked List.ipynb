# Traversing

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def display(self):
        if self.head is None:
            print("List is empty")
        else:
            p = self.head
            while p is not None:
                print(p.data)
                p = p.next
        
        
l = LinkedList()
l.head = Node(1)
l.display()
```

# Insert at Begin

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def display(self):
        if self.head is None:
            print("List is empty")
        else:
            p = self.head
            while p is not None:
                print(p.data, " ", end=" ")
                p = p.next
            print()
            
    def insertAtBeg(self, item):
        if self.head is None:
            self.head = Node(item)
        else:
            temp = Node(item)
            temp.next = self.head
            self.head = temp
        
        
l = LinkedList()
l.insertAtBeg(0)
l.insertAtBeg(-1)
l.insertAtBeg(-2)
l.insertAtBeg(-3)

l.display()
```

# Deletion of Last Node

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def display(self):
        if self.head is None:
            print("List is empty")
        else:
            p = self.head
            while p is not None:
                print(p.data, " ", end=" ")
                p = p.next
            print()
            
    def insertAtEnd(self, item):
        temp = Node(item)
        if self.head is None:
            self.head = temp
        else:
            p = self.head
            while p.next is not None:
                p = p.next
            p.next = temp
            
    def deleteLastNode(self):
        if self.head is None:
            return
        else:
            p = self.head
            if p.next is None:
                self.head = None
                return
            
            while p.next.next is not None:
                p = p.next
            p.next = None
            del p                
                        
l = LinkedList()

l.insertAtEnd(0)
l.insertAtEnd(1)
l.insertAtEnd(2)
l.insertAtEnd(3)

l.display()
l.deleteLastNode()
l.display()
```
