# Traversing

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def display(self):
        if self.head is None:
            print("List is empty")
        else:
            p = self.head
            while True:
                print(p.data," ",end="")
                p = p.next
                if(p == self.head):
                    break
        
        
l = LinkedList()
l.head = Node(1)
l.head.next = l.head
l.display()
```
