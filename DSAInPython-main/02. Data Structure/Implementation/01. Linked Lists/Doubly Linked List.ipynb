# Traversing

# Insert at Begin

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
        self.prev = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def display(self):
        if self.head is None:
            print("List is empty")
        else:
            p = self.head
            while p is not None:
                print(p.data, " ", end=" ")
                p = p.next
            print()
                
    def insertAtBeg(self, item):
        if self.head is None:
            temp = Node(item)
            self.head = temp
        else:
            temp = Node(item)
            temp.next = self.head
            self.head.prev = temp
            self.head = temp
        
        
l = LinkedList()
l.insertAtBeg(0)
l.insertAtBeg(-1)
l.insertAtBeg(-2)
l.insertAtBeg(-3)

l.display()
```
