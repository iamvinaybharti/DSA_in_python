# DFS (Iterative) - Adjacency list (Undirected/Unweighted)

```python
def add_node(v):
    if v in graph:
        print(v,"is already present in graph")
    else:
        graph[v] = []
        
def add_edge(v1, v2, cost=None):
    if v1 not in graph:
        print(v1,"is not present in graph")
    elif v2 not in graph:
        print(v2,"is not present in graph")
    else:
        if cost:
            list1 = [v2,cost]
            list2 = [v1,cost]
            graph[v1].append(list1)
            graph[v2].append(list2)
        else:
            graph[v1].append(v2)
            graph[v2].append(v1)
            
def DFS(node, graph):
    visited = set()
    if node not in graph:
        print("Node is not present")
        return
    stack = []
    stack.append(node)
    while stack:
        current = stack.pop()
        if current not in visited:
            print(current)
            visited.add(current)
            for i in graph[current]:
                stack.append(i)


graph = {}
add_node("A")
add_node("B")
add_node("C")
add_node("D")
add_node("E")
         


add_edge("A", "B")
add_edge("A", "C")
add_edge("B","D")
add_edge("D","E")

         
         
         

print("After Insertion:")
print(graph)  
print("After Traversal:")
DFS("A",graph)

```

# BFS (Iterative) - Adjacency list (Undirected/Unweighted)

```python
def add_node(v):
    if v in graph:
        print(v,"is already present in graph")
    else:
        graph[v] = []
        
def add_edge(v1, v2, cost=None):
    if v1 not in graph:
        print(v1,"is not present in graph")
    elif v2 not in graph:
        print(v2,"is not present in graph")
    else:
        if cost:
            list1 = [v2,cost]
            list2 = [v1,cost]
            graph[v1].append(list1)
            graph[v2].append(list2)
        else:
            graph[v1].append(v2)
            graph[v2].append(v1)
            
def BFS(node, graph):
    visited = set()
    if node not in graph:
        print("Node is not present")
        return
    queue = [node]
    while queue:
        current = queue.pop(0)
        print(current)
        visited.add(current)
        for i in graph[current]:
            if i not in visited:
                queue.append(i)


graph = {}
add_node("A")
add_node("B")
add_node("C")
add_node("D")
add_node("E")
         


add_edge("A", "B")
add_edge("A", "C")
add_edge("B","D")
add_edge("D","E")

         
         
         

print("After Insertion:")
print(graph)  
print("After Traversal:")
BFS("A",graph)

```
